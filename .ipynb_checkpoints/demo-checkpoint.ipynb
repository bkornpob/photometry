```python
from photometry.photometry import Photometry
```

# 0. Show

```python
from astropy.coordinates import SkyCoord
from astropy import wcs
from astropy import units as u
from astropy.io import fits

filename = './idlk01010_drz.fits'
skycoord = SkyCoord('2 58 10.52 -8 57 28.25', unit=(u.hourangle,u.deg))
w = wcs.WCS(fits.open(filename)[1].header)
```

```python
obj = Photometry(filename,skycoord,w)
obj.show2d(save=True,bbox=False)
```

# 1. BKG & SUB

```python
obj.show2d(save=False,bbox=True)
```

```python
obj.prep_polynomial2d()
obj.show3d(save=True)
```

```python
obj.fit_polynomial2d(norder=4)
obj.showsub(save=True)
```

```python
obj.savesub()
```

# 2. Photometry

```python
instrument = 'HST-WFC3-IR'
obj.photometry(instrument,save=True)
```
